```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Load data and exclude rows where Total Commands is 0
df = pd.read_csv("data-size-experiment.csv")
df = df[df['Total commands'] != 0]
```

```python
sns.set_theme(style="darkgrid")
plt.figure(figsize=(12, 6))
ax = sns.lineplot(x=(df['Total commands'] / 100000000).astype(int), y='Overall Throughput (cps)', hue='Policy', data=df)
plt.xticks(range(1, 14), labels=[str(i * 100) + 'M' for i in range(1, 14)])
ax.yaxis.set_major_formatter(FuncFormatter(lambda x, _: '{:,.0f}'.format(x)))
plt.xlabel('Total PUT Commands (x100,000,000)')
plt.title('Overall Throughput vs Number of PUTs up to 10GiB (1,342,177,280 key-value pairs)')
plt.legend(title='Policy')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
ax = sns.lineplot(x=(df['Total commands'] / 100000000).astype(int), y='Throughput (cps)', hue='Policy', data=df)
plt.xticks(range(1, 14), labels=[str(i * 100) + 'M' for i in range(1, 14)])
ax.yaxis.set_major_formatter(FuncFormatter(lambda x, _: '{:,.0f}'.format(x)))
plt.xlabel('Total PUT Commands (x100,000,000)')
plt.ylabel('Sliding Window Throughput (cps)')
plt.title('Sliding Window of 100,000,000 Throughput vs Number of PUTs up to 10GiB (1,342,177,280 key-value pairs)')
plt.legend(title='Policy')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
ax = sns.barplot(x=(df['Total commands'] / 100000000).astype(int), y='Faults Increase', hue='Policy', data=df)
plt.xticks(range(0, 13), labels=[str(i * 100) + 'M' for i in range(1, 14)])
ax.yaxis.set_major_formatter(FuncFormatter(lambda x, _: '{:,.0f}'.format(x)))
plt.xlabel('Total PUT Commands')
plt.ylabel('Page Faults in Sliding Window')
plt.title('Page Faults in Sliding Window of 100,000,000 Commands vs Number of PUTs')
plt.legend(title='Policy')
plt.show()
```
