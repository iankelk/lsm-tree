```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Load data and exclude rows where Total Commands is 0
df = pd.read_csv("data-size-experiment.csv")
df = df[df['Total commands'] != 0]

fig, axes = plt.subplots(1, 2, figsize=(24, 7))

ax1 = sns.lineplot(ax=axes[0], x=(df['Total commands'] / 100000000).astype(int), y='Throughput (cps)', hue='Policy', data=df)
axes[0].set_xticks(range(1, 14))
axes[0].set_xticklabels([str(i * 100) + 'M' for i in range(1, 14)], fontsize=14)
ax1.yaxis.set_major_formatter(FuncFormatter(lambda x, _: '{:,.0f}'.format(x)))
axes[0].set_xlabel('Total PUT Commands (x100,000,000)', fontsize=16)
axes[0].set_ylabel('Sliding Window Throughput (cps)', fontsize=16)
axes[0].set_title('Sliding Window of 100,000,000 Throughput vs Number of PUTs up to 10GiB (1,342,177,280 KV pairs)', fontsize=18)
axes[0].legend(title='Policy', fontsize=14, title_fontsize=14)
axes[0].tick_params(axis='y', labelsize=14)  # Increase y-axis tick label size

ax2 = sns.barplot(ax=axes[1], x=(df['Total commands'] / 100000000).astype(int), y='Faults Increase', hue='Policy', data=df)
axes[1].set_xticks(range(0, 13))
axes[1].set_xticklabels([str(i * 100) + 'M' for i in range(1, 14)], fontsize=14)
ax2.yaxis.set_major_formatter(FuncFormatter(lambda x, _: '{:,.0f}'.format(x)))
axes[1].set_xlabel('Total PUT Commands', fontsize=16)
axes[1].set_ylabel('Page Faults in Sliding Window', fontsize=16)
axes[1].set_title('Page Faults in Sliding Window of 100,000,000 Commands vs Number of PUTs', fontsize=18)
axes[1].legend(title='Policy', fontsize=14, title_fontsize=14)
axes[1].tick_params(axis='y', labelsize=14)  # Increase y-axis tick label size

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Load the CSV data
data = pd.read_csv('speedy.csv')

def comma_format(x, pos):
    return '{:,}'.format(int(x))

# Set up subplots
fig, axes = plt.subplots(1, 2, figsize=(20, 6))

# Plot 1: Overall Throughput vs Number of Commands
sns.lineplot(ax=axes[0], data=data, x='Total commands', y='Overall Throughput(cps)', hue='Buffer Size', palette='tab10')
axes[0].set_title('Overall Throughput vs Number of Commands')
axes[0].set_ylabel('Overall Throughput (cps)')
axes[0].legend(title='Buffer Size')
axes[0].xaxis.set_major_formatter(FuncFormatter(comma_format))
axes[0].yaxis.set_major_formatter(FuncFormatter(comma_format))

# Plot 2: Sliding Window of 200,000 commands Throughput vs Number of Commands
sns.lineplot(ax=axes[1], data=data, x='Total commands', y='Throughput(cps)', hue='Buffer Size', palette='tab10')
axes[1].set_title('Sliding Window of 200,000 commands Throughput vs Number of Commands')
axes[1].set_ylabel('Sliding Window Throughput (cps)')
axes[1].legend(title='Buffer Size')
axes[1].xaxis.set_major_formatter(FuncFormatter(comma_format))
axes[1].yaxis.set_major_formatter(FuncFormatter(comma_format))

# Adjust font sizes
plt.rcParams.update({'font.size': 14, 'axes.labelsize': 14, 'xtick.labelsize': 14, 'ytick.labelsize': 14, 'axes.titlesize': 16, 'legend.fontsize': 12})

plt.show()
```
