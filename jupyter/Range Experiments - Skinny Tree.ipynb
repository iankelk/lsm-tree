```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Read the CSV file
data = pd.read_csv("ranges_skinny_tree.csv")
```

```python
# First line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='Total commands', y='Overall Throughput', hue='Number of Threads', palette='tab10')
plt.title('Tiered Compaction: Overall Throughput vs Number of Commands per Second')
plt.ylabel('Overall Throughput (cps)')
plt.legend(title='Number of Threads')

# Format y-axis with commas
ax = plt.gca()
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))

# Format x-axis without scientific notation and with commas
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))

plt.ylim(data['Overall Throughput'].min(),250000)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 14, 'axes.labelsize': 16, 'xtick.labelsize': 14, 'ytick.labelsize': 14, 'axes.titlesize': 16, 'legend.fontsize': 12})

# Read the file
with open("range-sizes.txt", "r") as f:
    sizes = [int(line.strip()) for line in f.readlines()]

# Sort the sizes
sorted_sizes = sorted(sizes)

# Create the bar plot
plt.figure(figsize=(15, 6))
plt.bar(range(len(sorted_sizes)), sorted_sizes)

# Set the title and labels
plt.title("Range Sizes")
plt.xlabel("Index")
plt.ylabel("Number of Keys in Range")

# Show the plot
plt.show()
```
