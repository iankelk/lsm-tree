```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Read the CSV file
data = pd.read_csv("concurrent_4mil_puts_gets.csv")

# Custom legend labels
data['Clients'] = data['Number of Clients'].map({1: '1 client', 4: '4 clients'})

# Line plot 1
plt.figure(figsize=(10, 6))
plot1 = sns.lineplot(x='Total commands', y='Overall Throughput', hue='Policy', style='Clients', data=data)
plt.title('Overall Throughput vs Number of Commands per Second')

# Customize y-axis ticks to have commas
plot1.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

# Customize x-axis ticks to have "M" notation
plot1.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{x/1e6:,.0f}M'))

plt.legend(loc='best')
plt.show()

# Line plot 2
plt.figure(figsize=(10, 6))
plot2 = sns.lineplot(x='Total commands', y='Sliding Window Throughput', hue='Policy', style='Clients', data=data)
plt.title('Sliding Window of 1,000,000 commands Throughput vs Number of Commands per Second')

# Customize y-axis ticks to have commas
plot2.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

# Customize x-axis ticks to have "M" notation
plot2.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{x/1e6:,.0f}M'))

plt.legend(loc='best')
plt.show()
```
