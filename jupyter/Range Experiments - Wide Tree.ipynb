```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Read the CSV file
data = pd.read_csv("ranges_wide_tree.csv")
```

```python
# First line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='Total commands', y='Overall Throughput', hue='Number of Threads', palette='tab10')
plt.title('Tiered Compaction: Overall Throughput vs Number of Commands per Second')
plt.ylabel('Overall Throughput (cps)')
plt.legend(title='Number of Threads')

# Format y-axis with commas
ax = plt.gca()
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))

# Format x-axis without scientific notation and with commas
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.ylim(data['Overall Throughput'].min(),250000)
plt.xlim(350000, data['Total commands'].max())  # Set the x-axis bounds

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Read the CSV files
data1 = pd.read_csv("ranges_wide_tree.csv")
data2 = pd.read_csv("ranges_skinny_tree.csv")

# Set up the subplots
fig, axes = plt.subplots(1, 2, figsize=(20, 6))

# First plot
sns.lineplot(ax=axes[0], data=data1, x='Total commands', y='Overall Throughput', hue='Number of Threads', palette='tab10')
axes[0].set_title('Tiered Compaction Fanout = 10 (Wide Tree): 1M PUTs, 100 RANGEs')
axes[0].set_ylabel('Overall Throughput (cps)')
axes[0].legend(title='Threads', loc='lower left')

# Format y-axis with commas
axes[0].yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))

# Format x-axis without scientific notation and with commas
axes[0].xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))

axes[0].set_xlim(400000, data1['Total commands'].max())  # Set the x-axis bounds
axes[0].set_ylim(40000, 100000)  # Set the y-axis bounds

# Second plot
sns.lineplot(ax=axes[1], data=data2, x='Total commands', y='Overall Throughput', hue='Number of Threads', palette='tab10')
axes[1].set_title('Tiered Compaction Fanout = 2 (Skinny Tree): 1M PUTs, 100 RANGEs')
axes[1].set_ylabel('Overall Throughput (cps)')
axes[1].legend(title='Threads')

# Format y-axis with commas
axes[1].yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))

# Format x-axis without scientific notation and with commas
axes[1].xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))

axes[1].set_xlim(400000, data2['Total commands'].max())  # Set the x-axis bounds
axes[1].set_ylim(28000, 100000)  # Set the y-axis bounds

plt.show()
```
